# BrAPI Helper Demo

**Objective**:
The purpose of this notebook is to demonstrate how to retrieve data directly from a breeding management database without using any flat files, then perform basic visualizations with this data.

**Data**: In preparation for the demo we:
- Created an instance for the breeding (BMS) and genotype (GIGWA) data management systems (a.k.a databases).
- Created a demo dataset for a "toy" breeding trial on a `Maize` breeding program.
- Stored the genotype dataset in GIGWA.
- Stored the phenotypes on BMS for maize data.
